```python
plt.figure(figsize = (12,7))

Countries = ['USA', 'Brazil','Russia', 'Spain', 'UK', 'India']

total_deaths = [15628,654,8526,951,3251,159]

plt.bar(Countries,total_deaths,width = 0.5,align = 'center', color = 'cyan', edgecolor = 'red')

i = 1.0
j = 2000

for i in range(len(Countries)):
    plt.annotate(total_deaths[i], (-0.1 + i, total_deaths[i] + j))
    
plt.legend(labels =['total_deaths'])

plt.title('bar plot represents thge total deaths of the countries')

plt.xlabel('Countries')
plt.ylabel('Deaths')
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
plt.figure(figsize = (15,10))
totalDeaths = [15628,654,8526,951,3251,15955]
totalRecovery=[773480,325856,56452,323566,656535,146535]
activeCases=[133213,898463,12766,1313131,56313,135646]
Countries = ['USA', 'Brazil','Russia', 'Spain', 'UK', 'India']

X=np.arange(len(totalDeaths))

plt.bar(X,totalDeaths,color='black',width=0.25)
plt.bar(X+0.25,totalRecovery,color='g',width=0.25)
plt.bar(X+0.5,activeCases,color='b',width=0.25)

plt.legend(['Total Deaths','Total Recovery','Active Cases'])

plt.xticks([i+0.25 for i in range (6)],Countries)

plt.title("Bar plot representing the total deaths, total recovered cases ")

plt.xlabel('Countries')
plt.ylabel('Cases')

plt.savefig('4BarPlot.png')

plt.show()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt


Data={'Tasks':[300,500,700],
     'Task Type':['Task pending','Task ongoing','Task completed']
     }
df=DataFrame(Data)
df.set_index('Task Type',inplace=True)

df.plot.pie(y='Tasks',figsize=(10,10),autopct='%1.1f%%',startangle=90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels=['Civil','Electrical','Mechanical','Chemical']
sizes=[15,50,45,10]

fig,ax=plt.subplots()
ax.pie(sizes,labels=labels,autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Engineering Disciplines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels=['Civil','Electrical','Mechanical','Chemical']
sizes=[15,50,45,10]

explode=(0.1,0.1,0.1,0.4)

fig,ax=plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.1f%%',
      shadow=True,
      startangle=90)
ax.axis('equal')
ax.set_title('Engineering Disciplines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.bar(range(1,6),np.random.randint(1,20,5))
plt.title("2,2,1")

plt.subplot(2,2,2)
plt.bar(range(1,6),np.random.randint(1,20,5))
plt.title("2,2,2")

plt.subplot(2,2,3)
plt.scatter(range(1,6),np.random.randint(1,20,5),s=100,color="r")
plt.title("2,2,3")

plt.subplot(2,2,4)
plt.plot(range(1,6),np.random.randint(1,20,5),marker="o",color="g", linestyle="--")
plt.title("2,2,4")

```
